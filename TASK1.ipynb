```python
import psycopg2
import json
import sqlite3

conn = psycopg2.connect(database="boss", user="boss", password="sumanahappy1002", host="localhost", port=5432)


columns = [
    'tradedate',
    'instrument_id',
    'when_timestamp',
    'gamma',
    'vega',
    'theta'
]



cur = conn.cursor()
cur.execute('''DROP TABLE IF EXISTS trade_values''')
cur.execute('''CREATE TABLE trade_values
             (tradedate Date, instrument_id text, when_timestamp timestamp, gamma FLOAT, vega FLOAT, theta FLOAT  )''')



with open(r'/Users/boss/Downloads/json_events/valuedata.json') as f:
    for line in f:
        ndjson = json.loads(line)
        row = [ndjson[column] for column in columns]
         
        cur.execute('insert into trade_values values (%s,%s,%s,%s,%s,%s);', row)
        
        
cur.execute('SELECT * FROM trade_values LIMIT 100')        
print(cur.fetchall())

conn.commit()
cur.close()


```
